```python
import pandas as pd #data analysis
import matplotlib.pyplot as plt #visualizing
import seaborn as sns #visualizing
```

```python
#load the iris dataset
iris =pd.read_csv('/content/drive/MyDrive/visualizingdata/Iris.csv', skiprows=1 , names=['sepal_length','sepal_width','petal_length','petal_width','class'])
#check if the data is loaded
print(iris.head())
```

```python
print(iris.duplicated())
iris.info()
```

```python
print(iris)
```

```python
#colouring w.r.t to classes
colors={'Iris-setosa':'r','Iris-versicolor':'g','Iris-virginica':'b'}

#scatter plot
fig,axis =plt.subplots(figsize=(10,10))

#scatterring sepal length and sapal width
#scatter (x-data,y-data)
axis.scatter(iris['sepal_length'],iris['sepal_width'])
#coloring each points
for i in range(1,len(iris['sepal_length'])):
  axis.scatter(iris['sepal_length'][i],iris['sepal_width'][i],color =colors[iris['class'][i]])


axis.set_title('iris dataset')
axis.set_xlabel('sepal length')
axis.set_ylabel('sepal width')
```

```python
columns= iris.columns.drop(['class']) #choosing the columns to plot except class column
x_data=range(0,iris.shape[0])
fg,ax=plt.subplots(figsize =(10,10))

for i in columns:
  ax.plot(x_data , iris[i] ,label=i)

ax.set_title('iris dataset')
#ax.legend()
```
